```python
from ultralytics import YOLO

# Load an official or custom model
model = YOLO("yolo11n.pt")  # Load an official Detect model
model = YOLO("yolo11n-seg.pt")  # Load an official Segment model
model = YOLO("yolo11n-pose.pt")  # Load an official Pose model

# Perform tracking with the model
results = model.track("https://youtu.be/q3lX2p_Uy9I?si=u6eB25jhjrNSlFIF", show=True)  # Tracking with default tracker
results = model.track("https://youtu.be/q3lX2p_Uy9I?si=u6eB25jhjrNSlFIF", show=True, tracker="bytetrack.yaml")  # with ByteTrack
```
