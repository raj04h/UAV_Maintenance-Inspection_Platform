# Model Evaluation

```python
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

```python
model = YOLO(r"D:\Himanshu_ML\TE_UAV Projects\TE_1 UAV MRO\UAV_Dft\runs\detect\Deft_uav\weights\best.pt")
```

```python
results = model.val(data=r'D:\Himanshu_ML\TE_UAV Data\TE_defected\Combined\combined_dataset.yaml')
mAP50 = np.mean(results.box.map50)
mAP5095 = np.mean(results.box.map)
precision = np.mean(results.box.p)
recall = np.mean(results.box.r)
    
print(f"mAP@0.5: {mAP50:.4f}")
print(f"mAP@0.5:0.95: {mAP5095:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
```
